```python
import pandas as pd
import numpy as np
from math import ceil
```

```python
population = 150
sample = 15
k = ceil(population/sample)
print(k)
```

```python
r = np.random.randint(low=1, high= k + 1, size =1)
print(r)
```

```python
cumulative = r[0]
raffle = []
for i in range(sample):
  raffle.append(cumulative)
  cumulative += k
print(raffle)
```

```python
dataset = pd.read_csv('iris.csv')
dataset_final = dataset.loc[raffle]
dataset_final
```
